```python
# Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

```

```python
# utility functions

def check_shape(data):
    '''
    An util function to check data shape
    
    Args:
        df (pandas.DataFrame): Input DataFrame.
    '''
    shape = data.shape
    print(f"The data has {shape[0]} rows and {shape[1]} columns")
    

def print_unique_values(df, columns):
    """
    Utility function to print unique values of specified columns in a DataFrame.

    Args:
        df (pandas.DataFrame): Input DataFrame.
        columns (list): List of column names.
    """
    for column in columns:
        unique_values = df[column].unique().tolist()
        print(f"Unique values in column '{column}': {sorted(unique_values)}\n\n")

```

```python
# import data

# set path to data
path = "./Energy_Price_Prediction_/ml_model/data/clean_data.csv"
data_energy = pd.read_csv(path)

# print data head and tail 
display("First 5 rows of data")
display(data_energy.head())

display("Last 5 rows of data")
display(data_energy.tail())

```

```python
# Check the data shape
check_shape(data_energy)
```

```python
# check for missing values
plt.figure(figsize=(8, 6))
plt.title('Missing Data Visualization')
plt.xlabel('Columns')
plt.ylabel('Count of Missing Values')
data_energy.isna().sum().plot(kind="bar", color="steelblue")
plt.show()
```

The missing number of customers column is because the data was not caputred in earlier years, I will remove the rows.

```python
data_energy = data_energy.dropna()
check_shape(data_energy)
```

```python
# get unique values of state description and sector name
print_unique_values(data_energy, ['stateDescription', 'sectorName'])
```

Remove the row in the stateDescription that is U.S Total and all sectors row in the sectorName column

```python
# Remove rows where the "stateDescription" column is "allNation"
data_energy = data_energy[data_energy['stateDescription'] != 'U.S. Total']

# Remove rows where the "stateDescription" column is "allNation"
data_energy = data_energy[data_energy['sectorName'] != 'all sectors']

check_shape(data_energy)
```

```python
# check data info
data_energy.info()
```

```python
# check data description
data_energy.describe()
```

```python
# remove rows with price == 0
data_energy = data_energy[data_energy["price"] > 0]
```

```python
# remove rows with customers == 0
data_energy = data_energy[data_energy["customers"] > 0]
```

```python
# plot histogram of price
sns.histplot(data=data_energy, x = "price")
plt.title("Histogram of Price")
plt.show()
```

```python
sns.histplot(data=data_energy, x='customers')
```

```python
# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Encode the 'state' column
data_energy['state_encoded'] = label_encoder.fit_transform(data_energy['stateDescription'])
data_energy['sector_encoded'] = label_encoder.fit_transform(data_energy['sectorName'])

# plot corelation 
corr_matrix = data_energy.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# drop revenue and sales column
data_energy.drop(['revenue', 'sales'], inplace=True, axis =1)
check_shape(data_energy)
```

```python
# Combine "year" and "month" columns to create a new date column
data_energy['date'] = pd.to_datetime(data_energy[['year', 'month']].assign(day=1))

# Set the date column as the index
data_energy.set_index('date', inplace=True)

# Group by date, and calculate the mean price for each date
average_price = data_energy.groupby(['date'])['price'].mean()

# Plot the average price using a line plot
plt.figure(figsize=(10, 6))
average_price.plot(marker='o', linestyle='-')
plt.title('Average Price Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```

The plot shows the average price of energy over time, there seems to be an upward trend in the price of the energy. 
Let's check the price over time grouped by sectors.

```python
# Group by date and sector, and calculate the mean price for each group
average_price_sector = data_energy.groupby(['date', 'sectorName'])['price'].mean().reset_index()

# Plot the average price by sector for each date
plt.figure(figsize=(10, 6))
sns.lineplot(data=average_price_sector, x='date', y='price', hue='sectorName', marker='o')
plt.title('Average Price Over Time by Sector')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```

```python
# Filter the DataFrame for the year 2024
data_energy_2024 = data_energy[data_energy['year'] == 2024]

# Group by state and calculate the mean price for each state
average_price_by_state = data_energy_2024.groupby('stateDescription')['price'].mean().reset_index()

# Sort the DataFrame in descending order of the mean price
average_price_by_state = average_price_by_state.sort_values(by='price', ascending=False)

# Create a boxplot to visualize the distribution of prices across states in 2024
plt.figure(figsize=(10, 6))
sns.barplot(data=data_energy_2024, x='stateDescription', y='price', order=average_price_by_state['stateDescription'])
plt.title('Average Price by State in 2024')
plt.xlabel('State')
plt.ylabel('Price')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data_energy, x='customers', y='price')
plt.title('Scatter Plot of Consumer vs. Price')
plt.xlabel('Consumer')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```
